### DECISION TREE

```python
from sklearn.datasets import *
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor

import dtreeviz
import graphviz.backend as be
from IPython.display import Image, display_svg, SVG
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

import numpy as np
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
iris = load_iris()

X_train = iris.data
y_train = iris.target

clas = DecisionTreeClassifier()

# Fit the classifier to the training data
clas.fit(X_train, y_train)
```

```python
observation = [[ 5, 4, 3, 2]] # Predict observation's class
clas.predict(observation)
clas.predict_proba(observation)
```

### TREE VIZ

```python
plot_tree(clas)
```

```python
viz = dtreeviz.model(clas, X_train, y_train, feature_names=iris.feature_names, 
               class_names=["setosa", "versicolor", "virginica"])
viz
```
